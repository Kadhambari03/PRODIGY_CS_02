```python
import cv2
from google.colab.patches import cv2_imshow

def encrypt_image(image_path, output_path):
    img = cv2.imread(image_path)
    encrypted_img = 255 - img
    cv2.imwrite(output_path, encrypted_img)
    print("Encrypted image:")
    cv2_imshow(encrypted_img)

def decrypt_image(image_path, output_path):
    img = cv2.imread(image_path)
    decrypted_img = 255 - img
    cv2.imwrite(output_path, decrypted_img)
    print("Decrypted image:")
    cv2_imshow(decrypted_img)
encrypt_image("input.jpg", "encrypted.jpg")
```

```python
decrypt_image("encrypted.jpg", "decrypted.jpg")
```
